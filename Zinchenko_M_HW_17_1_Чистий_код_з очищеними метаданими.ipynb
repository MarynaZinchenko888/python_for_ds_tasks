````python
"""
Завдання 7. Аналіз e-commerce даних для двох країн (Germany vs France).
Використовуються бібліотеки pandas та sweetviz.
"""

# 7.1 Імпорт необхідних бібліотек
import warnings
# import pandas as pd завантажили в попередньому завданні, не повторюємось
import sweetviz as sv


# 2 Завантаження даних
def load_data(filepath: str) -> pd.DataFrame:
    """
    Завантажує дані з CSV-файлу з кодуванням ISO-8859-1.

    Args:
        filepath (str): шлях до CSV-файлу з даними.

    Returns:
        pd.DataFrame: датафрейм із завантаженими даними.
    """
    df = pd.read_csv(filepath, encoding='ISO-8859-1')
    return df


# 7.2 Обробка пропущених значень у CustomerID
def preprocess_customer_id(df: pd.DataFrame) -> pd.DataFrame:
    """
    Заповнює пропущені значення у колонці 'CustomerID' значенням '-1'
    та змінює тип даних на str (рядковий тип).

    Args:
        df (pd.DataFrame): вихідний датафрейм.

    Returns:
        pd.DataFrame: оновлений датафрейм.
    """
    df['CustomerID'] = df['CustomerID'].fillna(-1).astype(int).astype(str)
    return df


# 7.3 Побудова звіту Sweetviz
def create_sweetviz_report(df: pd.DataFrame) -> None:
    """
    Створює HTML-звіт Sweetviz, який порівнює дані по країнах Germany і France.

    Args:
        df (pd.DataFrame): датафрейм з даними e-commerce.
    """
    with warnings.catch_warnings(record=True):
        df_germany = df[df['Country'] == 'Germany']
        df_france = df[df['Country'] == 'France']

        # Створюємо звіт
        report = sv.compare([df_germany, "Germany"], [df_france, "France"])
        report.show_html('sweetviz_germany_france.html')
        print("✅ Звіт створено: sweetviz_germany_france.html")


# Основна функція
def main():
    """Основна функція виконання завдання 7."""
    filepath = r"/content/drive/MyDrive/ecommerce_data.csv"

    # 7.1. Завантаження даних
    df = load_data(filepath)
    print("✅ Дані завантажено успішно!")
    print(df.info())

    # 7.2. Обробка CustomerID
    df = preprocess_customer_id(df)
    print("✅ Колонку CustomerID очищено та змінено тип.")

    # 7.3. Побудова Sweetviz-звіту
    create_sweetviz_report(df)


# Виконання програми
if __name__ == "__main__":
    main()
````

````python
def calculate_variance(numbers):
    """
    Обчислює дисперсію числового списку оптимально через суму чисел та суму квадратів.

    Дисперсія визначається як (середнє квадратів) - (середнє)^2.

    Args:
        numbers (iterable of float/int): Ітерований об’єкт чисел.

    Returns:
        float: Дисперсія чисел у списку.
    """
    n = len(numbers)
    if n == 0:
        return 0.0
    sum_numbers = sum(numbers)
    sum_squares = sum(x * x for x in numbers)
    mean = sum_numbers / n
    return (sum_squares / n) - (mean ** 2)


input_list = [75, 80, 85, 90, 95]
print(calculate_variance(input_list))

````
